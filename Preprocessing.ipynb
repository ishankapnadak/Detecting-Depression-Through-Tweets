```python
import pandas as pd

random_df = pd.read_csv('random_tweets.csv', usecols=[1], names=['tweet'])
random_df.dropna(inplace=True)
random_df.reset_index(drop=True, inplace=True)
random_df.head(10)
```

```python
depressed_df = pd.read_csv('depressive_tweets_final.csv', usecols=[1])
depressed_df.head(10)
```

```python
import re
def fix(row):
    tweet = str(row['tweet'])
    tweet = re.sub("`", "'", tweet)
    return tweet

random_df['tweet'] = random_df.apply(fix, axis=1)
depressed_df['tweet'] = depressed_df.apply(fix, axis=1)
```

```python
import preprocessor as prep

def preprocess_tweet(row):
    tweet = str(row['tweet'])
    tweet = prep.clean(tweet)
    return tweet

random_df['tweet'] = random_df.apply(preprocess_tweet, axis=1)
depressed_df['tweet'] = depressed_df.apply(preprocess_tweet, axis=1)
```

```python
from gensim.parsing.preprocessing import remove_stopwords

def stopword_removal(row):
    tweet = str(row['tweet'])
    tweet = remove_stopwords(tweet)
    return tweet

random_df['tweet'] = random_df.apply(stopword_removal, axis=1)
depressed_df['tweet'] = depressed_df.apply(stopword_removal, axis=1)
```

```python
from nltk import PorterStemmer

def stem_tweet(row):
    tweet = str(row['tweet'])
    tweet = PorterStemmer().stem(tweet)
    return tweet

random_df['tweet'] = random_df.apply(stem_tweet, axis=1)
depressed_df['tweet'] = depressed_df.apply(stem_tweet, axis=1)
```
